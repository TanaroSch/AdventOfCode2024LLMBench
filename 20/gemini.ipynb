```python
from collections import deque

def solve():
    grid = []
    with open("input.txt", "r") as f:
        for line in f:
            grid.append(list(line.strip()))

    def find_start_end(grid):
        start = None
        end = None
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c] == 'S':
                    start = (r, c)
                elif grid[r][c] == 'E':
                    end = (r, c)
        return start, end

    def bfs(grid, start, end):
        q = deque([(start, 0)])
        visited = {start}
        while q:
            (r, c), dist = q.popleft()
            if (r, c) == end:
                return dist
            for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                nr, nc = r + dr, c + dc
                if 0 <= nr < len(grid) and 0 <= nc < len(grid[0]) and grid[nr][nc] != '#' and (nr, nc) not in visited:
                    visited.add((nr, nc))
                    q.append(((nr, nc), dist + 1))
        return -1

    def find_cheats(grid, start, end):
        cheats = []
        for r1 in range(len(grid)):
            for c1 in range(len(grid[0])):
                if grid[r1][c1] != '#':
                    for dr1, dc1 in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                        nr1, nc1 = r1 + dr1, c1 + dc1
                        if 0 <= nr1 < len(grid) and 0 <= nc1 < len(grid[0]):
                            for dr2, dc2 in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                                nr2, nc2 = nr1 + dr2, nc1 + dc2
                                if (dr1, dc1) != (-dr2, -dc2) and 0 <= nr2 < len(grid) and 0 <= nc2 < len(grid[0]) and grid[nr2][nc2] != '#':
                                    dist_saved = bfs(grid, start, (r1,c1)) + bfs(grid,(nr2, nc2),end) - (bfs(grid,start,end) - 2)
                                    if dist_saved > 0:
                                         cheats.append(((r1,c1),(nr2,nc2),dist_saved))

        return cheats

    start, end = find_start_end(grid)
    cheats = find_cheats(grid, start, end)

    count = 0
    for _, _, saved in cheats:
        if saved >= 100:
            count += 1
    print(count)

solve()
```

```python
from collections import deque

def solve():
    grid = []
    with open("input.txt", "r") as f:
        for line in f:
            grid.append(list(line.strip()))

    def find_start_end(grid):
        start = None
        end = None
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c] == 'S':
                    start = (r, c)
                elif grid[r][c] == 'E':
                    end = (r, c)
        return start, end

    def bfs(grid, start):
        distances = {}
        q = deque([(start, 0)])
        visited = {start}
        while q:
            (r, c), dist = q.popleft()
            distances[(r, c)] = dist
            for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                nr, nc = r + dr, c + dc
                if 0 <= nr < len(grid) and 0 <= nc < len(grid[0]) and grid[nr][nc] != '#' and (nr, nc) not in visited:
                    visited.add((nr, nc))
                    q.append(((nr, nc), dist + 1))
        return distances

    start, end = find_start_end(grid)
    start_distances = bfs(grid, start)
    end_distances = bfs(grid, end)
    shortest_path = start_distances.get(end, float('inf'))

    # dp[r][c] will store a set of (dist, cheat_start) tuples
    dp = {}
    q = deque([(start[0], start[1], 0, 0, start)])  # (r, c, cheat_time, dist, cheat_start)
    dp[(start[0], start[1])] = {(0, start)}  # Initial state: 0 distance, cheat_start is start itself
    
    cheats = set()

    while q:
        r, c, cheat_time, dist, cheat_start = q.popleft()

        if (r, c) in end_distances and cheat_time > 0:
            if dist + end_distances[(r, c)] < shortest_path:
                dist_saved = shortest_path - (dist + end_distances[(r, c)])
                if dist_saved >= 100:
                    cheats.add((cheat_start, (r, c)))
                    

        for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < len(grid) and 0 <= nc < len(grid[0]):
                
                new_dist = dist + 1

                if grid[nr][nc] != '#':
                    if (nr, nc) not in dp:
                        dp[(nr, nc)] = set()
                    
                    if (new_dist, cheat_start) not in dp[(nr, nc)]:
                        dp[(nr, nc)].add((new_dist, cheat_start))
                        q.append((nr, nc, cheat_time, new_dist, cheat_start))

                elif cheat_time < 20:
                    new_cheat_time = cheat_time + 1
                    
                    # Determine new cheat start
                    if cheat_time == 0:
                        new_cheat_start = (r, c)
                    else:
                        new_cheat_start = cheat_start

                    if (nr, nc) not in dp:
                        dp[(nr, nc)] = set()
                    
                    if (new_dist, new_cheat_start) not in dp[(nr, nc)]:
                        dp[(nr, nc)].add((new_dist, new_cheat_start))
                        q.append((nr, nc, new_cheat_time, new_dist, new_cheat_start))
    print(len(cheats))

solve()
```
