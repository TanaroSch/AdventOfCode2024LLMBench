```python
from collections import deque

def read_grid(filename):
    with open(filename, 'r') as f:
        grid = [line.strip() for line in f.readlines() if line.strip()]
    return grid

def find_positions(grid, target):
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == target:
                return x, y
    return None

def bfs(grid, start):
    rows, cols = len(grid), len(grid[0])
    d = [[-1 for _ in range(cols)] for _ in range(rows)]
    queue = deque()
    queue.append(start)
    d[start[1]][start[0]] = 0
    while queue:
        x, y = queue.popleft()
        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < cols and 0 <= ny < rows:
                if grid[ny][nx] != '#' and d[ny][nx] == -1:
                    d[ny][nx] = d[y][x] + 1
                    queue.append((nx, ny))
    return d

def main():
    grid = read_grid('input.txt')
    rows = len(grid)
    cols = len(grid[0]) if rows > 0 else 0

    start = find_positions(grid, 'S')
    end = find_positions(grid, 'E')

    d_S = bfs(grid, start)
    d_E = bfs(grid, end)

    T = d_S[end[1]][end[0]]
    print(f"Fastest time without cheating: {T} picoseconds")

    open_cells = []
    for y in range(rows):
        for x in range(cols):
            if grid[y][x] == '.' or grid[y][x] == 'S' or grid[y][x] == 'E':
                open_cells.append((x, y))

    count = 0
    directions = [(-1,0),(1,0),(0,-1),(0,1)]
    two_moves = [ (dx1+dx2, dy1+dy2) for dx1, dy1 in directions for dx2, dy2 in directions ]

    for u in open_cells:
        ux, uy = u
        for dx, dy in directions:
            vx = ux + dx
            vy = uy + dy
            if 0 <= vx < cols and 0 <= vy < rows and grid[vy][vx] in ['.', 'E']:
                v = (vx, vy)
                time_with_cheating = d_S[uy][ux] + 1 + d_E[vy][vx]
                time_saved = T - time_with_cheating
                if time_saved >= 100:
                    count += 1
        for move in two_moves:
            dx, dy = move
            vx = ux + dx
            vy = uy + dy
            if 0 <= vx < cols and 0 <= vy < rows and grid[vy][vx] in ['.', 'E']:
                v = (vx, vy)
                time_with_cheating = d_S[uy][ux] + 2 + d_E[vy][vx]
                time_saved = T - time_with_cheating
                if time_saved >= 100:
                    count += 1

    print(f"Number of cheats saving at least 100 picoseconds: {count}")

if __name__ == "__main__":
    main()
```

```python
from collections import deque

def read_grid(filename):
    with open(filename, 'r') as f:
        grid = [line.strip() for line in f.readlines() if line.strip()]
    return grid

def find_positions(grid, target):
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == target:
                return x, y
    return None

def bfs(grid, start):
    rows, cols = len(grid), len(grid[0])
    d = [[-1 for _ in range(cols)] for _ in range(rows)]
    queue = deque()
    queue.append(start)
    d[start[1]][start[0]] = 0
    while queue:
        x, y = queue.popleft()
        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < cols and 0 <= ny < rows:
                if grid[ny][nx] != '#' and d[ny][nx] == -1:
                    d[ny][nx] = d[y][x] + 1
                    queue.append((nx, ny))
    return d

def main():
    grid = read_grid('input.txt')
    rows = len(grid)
    cols = len(grid[0]) if rows > 0 else 0

    start = find_positions(grid, 'S')
    end = find_positions(grid, 'E')

    d_S = bfs(grid, start)
    d_E = bfs(grid, end)

    T = d_S[end[1]][end[0]]
    print(f"Fastest time without cheating: {T} picoseconds")

    open_cells = []
    for y in range(rows):
        for x in range(cols):
            if grid[y][x] in ['S', 'E', '.']:
                open_cells.append((x, y))

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    cheat_counts = [0] * 21  # Index represents cheat_duration from 0 to 20

    for u in open_cells:
        ux, uy = u
        for cheat_duration in range(1, 21):
            visited = [[False for _ in range(cols)] for _ in range(rows)]
            queue = deque()
            queue.append((ux, uy, 0))
            visited[uy][ux] = True
            while queue:
                x, y, steps = queue.popleft()
                if steps == cheat_duration:
                    continue
                for dx, dy in directions:
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < cols and 0 <= ny < rows:
                        if not visited[ny][nx]:
                            visited[ny][nx] = True
                            queue.append((nx, ny, steps + 1))
                            if grid[ny][nx] in ['S', 'E', '.']:
                                v = (nx, ny)
                                time_with_cheating = d_S[uy][ux] + cheat_duration + d_E[ny][nx]
                                time_saved = T - time_with_cheating
                                if time_saved >= 100:
                                    cheat_counts[cheat_duration] += 1

    total_cheats = sum(cheat_counts[1:21])
    print(f"Number of cheats saving at least 100 picoseconds: {total_cheats}")

if __name__ == "__main__":
    main()
```
