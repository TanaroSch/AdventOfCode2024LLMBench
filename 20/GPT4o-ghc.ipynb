```python
from collections import deque

def read_map(filename):
    with open(filename, 'r') as file:
        return [list(line.strip()) for line in file]

def find_start_end(racetrack):
    start = end = None
    for r, row in enumerate(racetrack):
        for c, val in enumerate(row):
            if val == 'S':
                start = (r, c)
            elif val == 'E':
                end = (r, c)
    return start, end

def bfs(racetrack, start, end):
    rows, cols = len(racetrack), len(racetrack[0])
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    queue = deque([(start, 0, False)])  # (position, time, cheated)
    visited = set([(start, False)])
    cheats = []

    while queue:
        (r, c), time, cheated = queue.popleft()
        
        if (r, c) == end:
            if cheated:
                cheats.append(time)
            continue
        
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                if racetrack[nr][nc] != '#' and ((nr, nc), cheated) not in visited:
                    visited.add(((nr, nc), cheated))
                    queue.append(((nr, nc), time + 1, cheated))
                elif racetrack[nr][nc] == '#' and not cheated:
                    for i in range(1, 3):
                        nnr, nnc = nr + dr * (i - 1), nc + dc * (i - 1)
                        if 0 <= nnr < rows and 0 <= nnc < cols and racetrack[nnr][nnc] != '#' and ((nnr, nnc), True) not in visited:
                            visited.add(((nnr, nnc), True))
                            queue.append(((nnr, nnc), time + i, True))
                            break

    return cheats

def count_cheats_saving_at_least(cheats, threshold):
    return sum(1 for cheat in cheats if cheat >= threshold)

def main():
    racetrack = read_map('input.txt')
    start, end = find_start_end(racetrack)
    cheats = bfs(racetrack, start, end)
    result = count_cheats_saving_at_least(cheats, 100)
    print(result)

if __name__ == "__main__":
    main()
```

```python
from collections import deque

def read_map(filename):
    with open(filename, 'r') as file:
        return [list(line.strip()) for line in file]

def find_positions(map_data):
    start = end = None
    for r, row in enumerate(map_data):
        for c, val in enumerate(row):
            if val == 'S':
                start = (r, c)
            elif val == 'E':
                end = (r, c)
    return start, end

def bfs(map_data, start, end, allow_cheat):
    rows, cols = len(map_data), len(map_data[0])
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    queue = deque([(start, 0, False)])  # (position, time, cheated)
    visited = set([(start, False)])
    min_time = float('inf')

    while queue:
        (r, c), time, cheated = queue.popleft()
        
        if (r, c) == end:
            min_time = min(min_time, time)
            continue
        
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                if map_data[nr][nc] != '#' and ((nr, nc), cheated) not in visited:
                    visited.add(((nr, nc), cheated))
                    queue.append(((nr, nc), time + 1, cheated))
                elif map_data[nr][nc] == '#' and not cheated and allow_cheat:
                    for dr2, dc2 in directions:
                        nnr, nnc = nr + dr2, nc + dc2
                        if 0 <= nnr < rows and 0 <= nnc < cols and map_data[nnr][nnc] != '#' and ((nnr, nnc), True) not in visited:
                            visited.add(((nnr, nnc), True))
                            queue.append(((nnr, nnc), time + 2, True))
    
    return min_time

def main():
    map_data = read_map('input.txt')
    start, end = find_positions(map_data)
    
    # Find the shortest path without cheating
    shortest_path_no_cheat = bfs(map_data, start, end, allow_cheat=False)
    
    # Find the shortest path with cheating
    shortest_path_with_cheat = bfs(map_data, start, end, allow_cheat=True)
    
    # Calculate the time saved by cheating
    time_saved = shortest_path_no_cheat - shortest_path_with_cheat
    
    # Count the number of cheats that save at least 100 picoseconds
    result = 1 if time_saved >= 100 else 0
    print(result)

if __name__ == "__main__":
    main()
```
