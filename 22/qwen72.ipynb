```python
def mix(secret, value):
    return secret ^ value

def prune(secret):
    return secret % 16777216

def generate_secret_numbers(initial_secret, count):
    secret = initial_secret
    for _ in range(count):
        secret = mix(secret, secret * 64)
        secret = prune(secret)
        secret = mix(secret, secret // 32)
        secret = prune(secret)
        secret = mix(secret, secret * 2048)
        secret = prune(secret)
    return secret

def main():
    with open('input.txt', 'r') as file:
        initial_secrets = [int(line.strip()) for line in file]

    total_sum = 0
    for initial_secret in initial_secrets:
        final_secret = generate_secret_numbers(initial_secret, 1999)  # 1999 steps to get the 2000th number
        total_sum += final_secret

    print(f"The sum of the 2000th secret number generated by each buyer is: {total_sum}")

if __name__ == "__main__":
    main()
```
