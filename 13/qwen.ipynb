```python
def read_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    machines = []
    for i in range(0, len(lines), 4):
        ax, ay = map(int, lines[i].split()[1][2:].split(','))
        bx, by = map(int, lines[i+1].split()[1][2:].split(','))
        px, py = map(int, lines[i+2].split()[1][2:].split(','))
        machines.append(((ax, ay), (bx, by), (px, py)))
    
    return machines

def min_tokens_to_win(ax, ay, bx, by, px, py):
    min_tokens = float('inf')
    for a in range(101):
        for b in range(101):
            if a * ax + b * bx == px and a * ay + b * by == py:
                tokens = a * 3 + b * 1
                if tokens < min_tokens:
                    min_tokens = tokens
    return min_tokens if min_tokens != float('inf') else None

def main():
    machines = read_input('input.txt')
    total_tokens = 0
    prizes_won = 0
    
    for ax, ay, bx, by, px, py in machines:
        tokens = min_tokens_to_win(ax, ay, bx, by, px, py)
        if tokens is not None:
            total_tokens += tokens
            prizes_won += 1
    
    print(f"Prizes won: {prizes_won}")
    print(f"Minimum tokens required: {total_tokens}")

if __name__ == "__main__":
    main()
```

```python
def parse_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    machines = []
    i = 0
    while i < len(lines):
        # Parse Button A
        ax, ay = map(int, lines[i].split('+')[1].split(', Y+'))
        # Parse Button B
        bx, by = map(int, lines[i+1].split('+')[1].split(', Y+'))
        # Parse Prize
        px, py = map(int, lines[i+2].split('=')[1].split(', Y='))
        machines.append(((ax, ay), (bx, by), (px, py)))
        i += 4
    return machines

def find_min_cost(ax, ay, bx, by, px, py):
    min_cost = float('inf')
    for a in range(101):
        for b in range(101):
            if a * ax + b * bx == px and a * ay + b * by == py:
                cost = a * 3 + b * 1
                if cost < min_cost:
                    min_cost = cost
    return min_cost if min_cost != float('inf') else None

def main():
    machines = parse_input('input.txt')
    total_min_cost = 0
    for (ax, ay), (bx, by), (px, py) in machines:
        min_cost = find_min_cost(ax, ay, bx, by, px, py)
        if min_cost is not None:
            total_min_cost += min_cost
    print(total_min_cost)

if __name__ == "__main__":
    main()
```
