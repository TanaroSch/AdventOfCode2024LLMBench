```python
def parse_input(file_path):
    machines = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for i in range(0, len(lines), 4):
            a_line = lines[i].strip()
            b_line = lines[i+1].strip()
            prize_line = lines[i+2].strip()
            
            a_x = int(a_line.split()[2][2:].strip(','))
            a_y = int(a_line.split()[3][2:])
            b_x = int(b_line.split()[2][2:].strip(','))
            b_y = int(b_line.split()[3][2:])
            prize_x = int(prize_line.split()[1][2:].strip(','))
            prize_y = int(prize_line.split()[2][2:])
            
            machines.append(((a_x, a_y), (b_x, b_y), (prize_x, prize_y)))
    return machines

def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def find_min_tokens(machines):
    min_tokens = 0
    for a, b, prize in machines:
        a_x, a_y = a
        b_x, b_y = b
        prize_x, prize_y = prize
        
        prize_x += 10000000000000
        prize_y += 10000000000000
        
        gcd_x, x, y = extended_gcd(a_x, b_x)
        if prize_x % gcd_x != 0:
            continue
        
        gcd_y, x, y = extended_gcd(a_y, b_y)
        if prize_y % gcd_y != 0:
            continue
        
        # Scale the prize coordinates
        prize_x //= gcd_x
        prize_y //= gcd_y
        
        a_x //= gcd_x
        b_x //= gcd_x
        a_y //= gcd_y
        b_y //= gcd_y
        
        # Use the Extended Euclidean Algorithm to find a solution
        gcd, x0, y0 = extended_gcd(a_x, b_x)
        if prize_x % gcd != 0:
            continue
        
        x0 *= prize_x // gcd
        y0 *= prize_x // gcd
        
        gcd, x1, y1 = extended_gcd(a_y, b_y)
        if prize_y % gcd != 0:
            continue
        
        x1 *= prize_y // gcd
        y1 *= prize_y // gcd
        
        # Find the minimum cost
        found = False
        min_cost = float('inf')
        
        for k in range(-100, 101):
            i = x0 + k * (b_x // gcd)
            j = y0 - k * (a_x // gcd)
            if i >= 0 and j >= 0:
                cost = i * 3 + j
                if cost < min_cost:
                    min_cost = cost
                    found = True
        
        if found:
            min_tokens += min_cost
    
    return min_tokens

machines = parse_input('input.txt')
result = find_min_tokens(machines)
print(result)
```
