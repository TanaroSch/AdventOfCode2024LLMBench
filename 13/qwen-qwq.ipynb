```python
import re

with open('input.txt', 'r') as f:
    lines = f.readlines()

machines = []
i = 0
while i < len(lines):
    if not lines[i].strip():
        i += 1
        continue  # Skip empty lines
    machine = []
    # Read line1: Button A: X+a, Y+c
    line1 = lines[i].strip()
    numbers = re.findall('\d+', line1)
    if len(numbers) < 2:
        print("Insufficient numbers in line1")
        i += 1
        continue
    a, c = int(numbers[0]), int(numbers[1])
    machine.append(a)
    machine.append(c)
    
    # Read line2: Button B: X+b, Y+d
    i += 1
    if i >= len(lines):
        print("Missing line2 for this machine")
        break
    line2 = lines[i].strip()
    numbers = re.findall('\d+', line2)
    if len(numbers) < 2:
        print("Insufficient numbers in line2")
        i += 1
        continue
    b, d = int(numbers[0]), int(numbers[1])
    machine.append(b)
    machine.append(d)
    
    # Read line3: Prize: X=x, Y=y
    i += 1
    if i >= len(lines):
        print("Missing line3 for this machine")
        break
    line3 = lines[i].strip()
    numbers = re.findall('\d+', line3)
    if len(numbers) < 2:
        print("Insufficient numbers in line3")
        i += 1
        continue
    x, y = int(numbers[0]), int(numbers[1])
    machine.append(x)
    machine.append(y)
    
    machines.append(machine)
    i += 1  # Move to the next set of lines

# Now process each machine
total_cost = 0
for machine in machines:
    a, c, b, d, x, y = machine
    min_cost = float('inf')
    for A in range(101):
        for B in range(101):
            if a*A + b*B == x and c*A + d*B == y:
                cost = 3*A + 1*B
                if cost < min_cost:
                    min_cost = cost
    if min_cost != float('inf'):
        total_cost += min_cost
print(total_cost)
```

```python
import re
from math import gcd

def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    g, x, y = extended_gcd(b % a, a)
    return g, y - (b // a) * x, x

def solve_diophantine(a, b, c):
    g, x, y = extended_gcd(a, b)
    if c % g != 0:
        return None
    x0 = x * (c // g)
    y0 = y * (c // g)
    return x0, y0, g

def find_min_tokens(a, c, b, d, x, y):
    # Solve a*A + b*B = x and c*A + d*B = y
    # First, solve a*A + b*B = x
    g_ab = gcd(a, b)
    if x % g_ab != 0:
        return float('inf')
    A0, B0, g = solve_diophantine(a, b, x)
    if A0 is None:
        return float('inf')
    # Now, check if c*A + d*B = y
    # Substitute A = A0 + k*(b//g), B = B0 - k*(a//g)
    # So, c*(A0 + k*(b//g)) + d*(B0 - k*(a//g)) = y
    # c*A0 + c*k*(b//g) + d*B0 - d*k*(a//g) = y
    # (c*(b//g) - d*(a//g))*k = y - c*A0 - d*B0
    cb_d_a = c * (b // g) - d * (a // g)
    rhs = y - c * A0 - d * B0
    if cb_d_a == 0:
        if rhs != 0:
            return float('inf')
        else:
            # Infinite solutions, find the one with minimal tokens
            # Tokens = 3*A + B
            # A = A0 + k*(b//g)
            # B = B0 - k*(a//g)
            # Find k such that A >= 0 and B >= 0
            min_tokens = float('inf')
            for k in range(-1000000, 1000000):
                A = A0 + k * (b // g)
                B = B0 - k * (a // g)
                if A >= 0 and B >= 0:
                    tokens = 3 * A + B
                    if tokens < min_tokens:
                        min_tokens = tokens
            return min_tokens
    else:
        if rhs % cb_d_a != 0:
            return float('inf')
        k = rhs // cb_d_a
        A = A0 + k * (b // g)
        B = B0 - k * (a // g)
        if A >= 0 and B >= 0:
            return 3 * A + B
        else:
            return float('inf')

with open('input.txt', 'r') as f:
    lines = f.readlines()

machines = []
for i in range(0, len(lines), 3):
    machine = []
    line1 = lines[i].strip()
    numbers = re.findall(r'\d+', line1)
    if len(numbers) < 2:
        continue
    a, c = int(numbers[0]), int(numbers[1])
    line2 = lines[i+1].strip()
    numbers = re.findall(r'\d+', line2)
    if len(numbers) < 2:
        continue
    b, d = int(numbers[0]), int(numbers[1])
    line3 = lines[i+2].strip()
    numbers = re.findall(r'\d+', line3)
    if len(numbers) < 2:
        continue
    x, y = int(numbers[0]) + 10000000000000, int(numbers[1]) + 10000000000000
    machine.append(a)
    machine.append(c)
    machine.append(b)
    machine.append(d)
    machine.append(x)
    machine.append(y)
    machines.append(machine)

total_cost = 0
for machine in machines:
    a, c, b, d, x, y = machine
    min_cost = find_min_tokens(a, c, b, d, x, y)
    if min_cost != float('inf'):
        total_cost += min_cost
print(total_cost)
```

```python
import re
from math import gcd

def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    g, x, y = extended_gcd(b % a, a)
    return g, y - (b // a) * x, x

def solve_diophantine(a, b, c):
    g, x, y = extended_gcd(a, b)
    if c % g != 0:
        return None
    x0 = x * (c // g)
    y0 = y * (c // g)
    return x0, y0, g

def find_min_tokens(a, c, b, d, x, y):
    # Solve for A and B in the system:
    # a*A + b*B = x
    # c*A + d*B = y
    det = a*d - b*c
    if det == 0:
        # No unique solution
        return float('inf')
    
    # Using Cramer's rule
    det_x = x*d - b*y
    det_y = a*y - x*d
    if det_x % det != 0 or det_y % det != 0:
        # No integer solution
        return float('inf')
    
    A = det_x // det
    B = det_y // det
    if A < 0 or B < 0:
        # Negative presses are not allowed
        return float('inf')
    
    # Calculate cost
    cost = 3*A + 1*B
    return cost

with open('input.txt', 'r') as f:
    lines = f.readlines()

machines = []
num_lines = len(lines)
i = 0
while i < num_lines:
    machine = []
    line1 = lines[i].strip()
    print(f"Processing machine starting at line {i}")
    print(f"Line 1: {line1}")
    numbers = re.findall(r'\d+', line1)
    if len(numbers) < 2:
        print("Insufficient numbers in line1")
        i += 3
        continue
    a, c = int(numbers[0]), int(numbers[1])
    
    if i+1 >= num_lines:
        print("Missing line2")
        break
    line2 = lines[i+1].strip()
    print(f"Line 2: {line2}")
    numbers = re.findall(r'\d+', line2)
    if len(numbers) < 2:
        print("Insufficient numbers in line2")
        i += 3
        continue
    b, d = int(numbers[0]), int(numbers[1])
    
    if i+2 >= num_lines:
        print("Missing line3")
        break
    line3 = lines[i+2].strip()
    print(f"Line 3: {line3}")
    numbers = re.findall(r'\d+', line3)
    if len(numbers) < 2:
        print("Insufficient numbers in line3")
        i += 3
        continue
    x, y = int(numbers[0]) + 10000000000000, int(numbers[1]) + 10000000000000
    machine.append(a)
    machine.append(c)
    machine.append(b)
    machine.append(d)
    machine.append(x)
    machine.append(y)
    machines.append(machine)
    i += 3

total_cost = 0
for machine in machines:
    a, c, b, d, x, y = machine
    min_cost = find_min_tokens(a, c, b, d, x, y)
    if min_cost != float('inf'):
        total_cost += min_cost
print(total_cost)
```
