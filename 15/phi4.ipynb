```python
def read_input(file_path):
    with open(file_path, 'r') as file:
        lines = [line.strip() for line in file]
    
    grid = []
    movements = []
    reading_movements = False
    
    for line in lines:
        if line == "":
            reading_movements = True
            continue
        
        if not reading_movements:
            grid.append(list(line))
        else:
            movements.extend(list(line))
    
    return grid, movements

def find_positions(grid):
    robot_pos = None
    box_positions = []
    
    for r in range(len(grid)):
        for c in range(len(grid[r])):
            if grid[r][c] == '@':
                robot_pos = (r, c)
            elif grid[r][c] == 'O':
                box_positions.append((r, c))
    
    return robot_pos, box_positions

def move_robot_and_boxes(grid, movements):
    direction_map = {
        '^': (-1, 0),
        'v': (1, 0),
        '<': (0, -1),
        '>': (0, 1)
    }
    
    def is_within_bounds(r, c):
        return 0 <= r < len(grid) and 0 <= c < len(grid[0])
    
    robot_pos, box_positions = find_positions(grid)
    
    for move in movements:
        dr, dc = direction_map[move]
        new_r, new_c = robot_pos[0] + dr, robot_pos[1] + dc
        
        if is_within_bounds(new_r, new_c) and grid[new_r][new_c] != '#':
            # Check if there's a box in the way
            if (new_r, new_c) in box_positions:
                new_box_r, new_box_c = new_r + dr, new_c + dc
                
                if is_within_bounds(new_box_r, new_box_c) and grid[new_box_r][new_box_c] == '.':
                    # Move the box
                    grid[robot_pos[0]][robot_pos[1]] = '.'
                    grid[new_r][new_c] = 'O'
                    grid[new_box_r][new_box_c] = '@'
                    
                    robot_pos = (new_box_r, new_box_c)
                    box_positions.remove((new_r, new_c))
                    box_positions.append((new_box_r, new_box_c))
            else:
                # Move the robot
                grid[robot_pos[0]][robot_pos[1]] = '.'
                grid[new_r][new_c] = '@'
                robot_pos = (new_r, new_c)
    
    return box_positions

def calculate_gps_sum(box_positions):
    gps_sum = 0
    
    for r, c in box_positions:
        gps_sum += 100 * r + c
    
    return gps_sum

def main(file_path):
    grid, movements = read_input(file_path)
    final_box_positions = move_robot_and_boxes(grid, movements)
    result = calculate_gps_sum(final_box_positions)
    
    print(result)

# Assuming the input file is named 'input.txt'
main('input.txt')
```

```python
def read_input(file_path):
    with open(file_path, 'r') as file:
        lines = [line.strip() for line in file]
    
    grid = []
    movements = []
    reading_movements = False
    
    for line in lines:
        if line == "":
            reading_movements = True
            continue
        
        if not reading_movements:
            grid.append(list(line))
        else:
            movements.extend(list(line))
    
    return grid, movements

def find_positions(grid):
    robot_pos = None
    box_positions = []
    
    for r in range(len(grid)):
        for c in range(len(grid[r])):
            if grid[r][c] == '@':
                robot_pos = (r, c)
            elif grid[r][c] == 'O':
                box_positions.append((r, c))
    
    return robot_pos, box_positions

def print_grid(grid):
    for row in grid:
        print("".join(row))
    print()

def move_robot_and_boxes(grid, movements):
    direction_map = {
        '^': (-1, 0),
        'v': (1, 0),
        '<': (0, -1),
        '>': (0, 1)
    }
    
    def is_within_bounds(r, c):
        return 0 <= r < len(grid) and 0 <= c < len(grid[0])
    
    robot_pos, box_positions = find_positions(grid)
    
    print("Initial Grid:")
    print_grid(grid)
    print(f"Initial Robot Position: {robot_pos}")
    print(f"Initial Box Positions: {box_positions}\n")
    
    for move in movements:
        dr, dc = direction_map[move]
        new_r, new_c = robot_pos[0] + dr, robot_pos[1] + dc
        
        if is_within_bounds(new_r, new_c):
            cell_content = grid[new_r][new_c]
            
            if cell_content == '.' or cell_content == '*':
                # Move the robot
                print(f"Moving Robot to empty space: {(new_r, new_c)}")
                grid[robot_pos[0]][robot_pos[1]] = '.'
                grid[new_r][new_c] = '@'
                robot_pos = (new_r, new_c)
                
            elif cell_content == 'O':
                # Check if the box can be pushed
                new_box_r, new_box_c = new_r + dr, new_c + dc
                
                if is_within_bounds(new_box_r, new_box_c) and grid[new_box_r][new_box_c] in ('.', '*'):
                    print(f"Pushing Box from {(new_r, new_c)} to {(new_box_r, new_box_c)}")
                    grid[robot_pos[0]][robot_pos[1]] = '.'
                    grid[new_r][new_c] = '@'
                    grid[new_box_r][new_box_c] = 'O' if grid[new_box_r][new_box_c] == '.' else 'X'
                    
                    robot_pos = (new_r, new_c)
                    box_positions.remove((new_r, new_c))
                    box_positions.append((new_box_r, new_box_c))
                else:
                    print(f"Failed to push Box from {(new_r, new_c)} due to obstacle or boundary")
            
            elif cell_content == 'X':
                # Move robot onto target
                print(f"Moving Robot to target: {(new_r, new_c)}")
                grid[robot_pos[0]][robot_pos[1]] = '.'
                grid[new_r][new_c] = '@'
                robot_pos = (new_r, new_c)
        
        else:
            print(f"Move {move} is out of bounds.")
        
        print("Grid after move:")
        print_grid(grid)
    
    return box_positions

def calculate_gps_sum(box_positions):
    gps_sum = 0
    
    for r, c in box_positions:
        gps_sum += 100 * r + c
    
    return gps_sum

def main(file_path):
    grid, movements = read_input(file_path)
    final_box_positions = move_robot_and_boxes(grid, movements)
    result = calculate_gps_sum(final_box_positions)
    
    print(f"Final Box Positions: {final_box_positions}")
    print(f"Total Sum of Final Box Coordinates: {result}")

# Example usage:
main('input.txt')
```

```python
def read_input(filename):
    with open(filename, 'r') as file:
        lines = file.read().strip().split('\n')
    
    # Split input into grid and moves
    empty_line_index = lines.index('')
    warehouse_grid = [list(line) for line in lines[:empty_line_index]]
    movements = lines[empty_line_index + 1]
    
    return warehouse_grid, movements

def find_robot_position(grid):
    for y, row in enumerate(grid):
        for x, cell in enumerate(row):
            if cell == '@':
                return x, y
    return None

def move_box(grid, from_pos, to_pos):
    fx, fy = from_pos
    tx, ty = to_pos
    grid[ty][tx] = 'O'  # Move box to new position
    grid[fy][fx] = '.'   # Clear old box position

def simulate_movements(warehouse_grid, movements):
    directions = {
        '<': (0, -1),
        '>': (0, 1),
        '^': (-1, 0),
        'v': (1, 0)
    }
    
    robot_pos = find_robot_position(warehouse_grid)
    rows, cols = len(warehouse_grid), len(warehouse_grid[0])
    
    for move in movements:
        dx, dy = directions[move]
        new_robot_x = robot_pos[0] + dx
        new_robot_y = robot_pos[1] + dy
        
        # Check if the new position is within bounds and not a wall
        if 0 <= new_robot_x < cols and 0 <= new_robot_y < rows:
            target_cell = warehouse_grid[new_robot_y][new_robot_x]
            
            if target_cell == '.':
                # Move robot to empty space
                warehouse_grid[robot_pos[1]][robot_pos[0]] = '.'
                robot_pos = (new_robot_x, new_robot_y)
                warehouse_grid[new_robot_y][new_robot_x] = '@'
                
            elif target_cell == 'O':
                # Attempt to push the box
                new_box_x = new_robot_x + dx
                new_box_y = new_robot_y + dy
                
                if 0 <= new_box_x < cols and 0 <= new_box_y < rows:
                    beyond_box = warehouse_grid[new_box_y][new_box_x]
                    
                    if beyond_box == '.' or beyond_box == '@':
                        move_box(warehouse_grid, (new_robot_x, new_robot_y), (new_box_x, new_box_y))
                        # Move robot to the box's previous position
                        warehouse_grid[robot_pos[1]][robot_pos[0]] = '.'
                        robot_pos = (new_robot_x, new_robot_y)
                        warehouse_grid[new_robot_y][new_robot_x] = '@'

def calculate_gps_sum(warehouse_grid):
    gps_sum = 0
    rows, cols = len(warehouse_grid), len(warehouse_grid[0])
    
    for y in range(rows):
        for x in range(cols):
            if warehouse_grid[y][x] == 'O':
                gps_coordinate = 100 * y + x
                gps_sum += gps_coordinate
                
    return gps_sum

def main():
    warehouse_grid, movements = read_input('input.txt')
    simulate_movements(warehouse_grid, movements)
    result = calculate_gps_sum(warehouse_grid)
    print(result)

if __name__ == "__main__":
    main()
```
