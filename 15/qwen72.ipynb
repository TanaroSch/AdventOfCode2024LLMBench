```python
def read_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    # Parse the warehouse map
    warehouse = []
    for line in lines:
        if line.strip() and line.strip()[0] != '<':
            warehouse.append(list(line.strip()))
        else:
            break
    
    # Parse the moves
    moves = ''.join(line.strip() for line in lines if line.strip() and line.strip()[0] == '<')
    
    return warehouse, moves

def find_robot_position(warehouse):
    for i, row in enumerate(warehouse):
        for j, cell in enumerate(row):
            if cell == '@':
                return i, j
    return None

def move_robot(warehouse, robot_pos, move):
    i, j = robot_pos
    di, dj = {'^': (-1, 0), 'v': (1, 0), '<': (0, -1), '>': (0, 1)}[move]
    ni, nj = i + di, j + dj
    
    if warehouse[ni][nj] == '#':
        return robot_pos  # Wall, no movement
    
    if warehouse[ni][nj] == 'O':
        # Push the box
        nni, nnj = ni + di, nj + dj
        if warehouse[nni][nnj] == '#':
            return robot_pos  # Wall, no movement
        if warehouse[nni][nnj] == 'O':
            return robot_pos  # Another box, no movement
        warehouse[nni][nnj] = 'O'
        warehouse[ni][nj] = '.'
    
    warehouse[ni][nj] = '@'
    warehouse[i][j] = '.'
    return ni, nj

def calculate_gps(warehouse):
    total_gps = 0
    for i, row in enumerate(warehouse):
        for j, cell in enumerate(row):
            if cell == 'O':
                total_gps += 100 * i + j
    return total_gps

def simulate_moves(warehouse, moves):
    robot_pos = find_robot_position(warehouse)
    for move in moves:
        robot_pos = move_robot(warehouse, robot_pos, move)
    return warehouse

def main():
    warehouse, moves = read_input('input.txt')
    final_warehouse = simulate_moves(warehouse, moves)
    total_gps = calculate_gps(final_warehouse)
    print(f"The sum of all boxes' GPS coordinates is: {total_gps}")

if __name__ == "__main__":
    main()
```
