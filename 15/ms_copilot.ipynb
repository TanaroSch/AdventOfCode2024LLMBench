```python
def read_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    return lines

def parse_input(lines):
    map_lines = []
    movements = []
    reading_map = True

    for line in lines:
        line = line.strip()
        if reading_map:
            if line.startswith('#'):
                map_lines.append(line)
            else:
                reading_map = False
                movements.append(line)
        else:
            movements.append(line)

    # Join movement lines and remove any potential newlines within the moves
    movements = ''.join(movements).replace('\n', '')

    return map_lines, movements

def find_robot_and_boxes(map_lines):
    robot_pos = None
    boxes = []

    for y, line in enumerate(map_lines):
        for x, char in enumerate(line):
            if char == '@':
                robot_pos = (x, y)
            elif char == 'O':
                boxes.append((x, y))

    return robot_pos, boxes

def is_valid_position(x, y, map_lines):
    return 0 <= y < len(map_lines) and 0 <= x < len(map_lines[y]) and map_lines[y][x] != '#'

def move_robot_and_boxes(robot_pos, boxes, movements, map_lines):
    directions = {'<': (-1, 0), '>': (1, 0), '^': (0, -1), 'v': (0, 1)}

    for move in movements:
        dx, dy = directions[move]
        new_robot_pos = (robot_pos[0] + dx, robot_pos[1] + dy)

        if is_valid_position(new_robot_pos[0], new_robot_pos[1], map_lines):
            # Check if there's a box at the new robot position
            if new_robot_pos in boxes:
                new_box_pos = (new_robot_pos[0] + dx, new_robot_pos[1] + dy)
                if is_valid_position(new_box_pos[0], new_box_pos[1], map_lines) and new_box_pos not in boxes:
                    # Move the box
                    boxes.remove(new_robot_pos)
                    boxes.append(new_box_pos)
                    robot_pos = new_robot_pos
            else:
                robot_pos = new_robot_pos

    return robot_pos, boxes

def calculate_gps_coordinates(boxes, map_lines):
    gps_sum = 0
    for x, y in boxes:
        gps_sum += 100 * y + x
    return gps_sum

def main():
    lines = read_input('input.txt')
    map_lines, movements = parse_input(lines)
    robot_pos, boxes = find_robot_and_boxes(map_lines)
    robot_pos, boxes = move_robot_and_boxes(robot_pos, boxes, movements, map_lines)
    gps_sum = calculate_gps_coordinates(boxes, map_lines)
    print(gps_sum)

if __name__ == '__main__':
    main()
```
