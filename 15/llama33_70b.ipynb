```python
def parse_input(file_name):
    with open(file_name, 'r') as f:
        lines = f.readlines()
    warehouse_map = [list(line.strip()) for line in lines[:-1]]
    moves = lines[-1].strip().replace('\n', '')
    return warehouse_map, moves

def find_robot(warehouse_map):
    for y, row in enumerate(warehouse_map):
        for x, cell in enumerate(row):
            if cell == '@':
                return x, y

def is_valid_move(warehouse_map, x, y):
    return 0 <= x < len(warehouse_map[0]) and 0 <= y < len(warehouse_map)

def move_robot(warehouse_map, x, y, dx, dy):
    new_x, new_y = x + dx, y + dy
    if not is_valid_move(warehouse_map, new_x, new_y) or warehouse_map[new_y][new_x] == '#':
        return x, y
    if warehouse_map[new_y][new_x] == 'O':
        new_new_x, new_new_y = new_x + dx, new_y + dy
        if not is_valid_move(warehouse_map, new_new_x, new_new_y) or warehouse_map[new_new_y][new_new_x] == '#':
            return x, y
        warehouse_map[new_y][new_x], warehouse_map[new_new_y][new_new_x] = '.', 'O'
    warehouse_map[y][x], warehouse_map[new_y][new_x] = '.', '@'
    return new_x, new_y

def calculate_gps(warehouse_map):
    gps_sum = 0
    for y, row in enumerate(warehouse_map):
        for x, cell in enumerate(row):
            if cell == 'O':
                gps_sum += 100 * y + x
    return gps_sum

def main():
    warehouse_map, moves = parse_input('input.txt')
    x, y = find_robot(warehouse_map)
    move_dict = {'<': (-1, 0), '>': (1, 0), '^': (0, -1), 'v': (0, 1)}
    for move in moves:
        dx, dy = move_dict[move]
        x, y = move_robot(warehouse_map, x, y, dx, dy)
    print(calculate_gps(warehouse_map))

if __name__ == '__main__':
    main()
```
