```python
def read_input(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    warehouse = []
    moves = ""
    for line in lines:
        if line.startswith('#'):
            warehouse.append(line.strip())
        else:
            moves += line.strip()
    
    return warehouse, moves

def find_robot_and_boxes(warehouse):
    robot_pos = None
    boxes = set()
    for r, row in enumerate(warehouse):
        for c, char in enumerate(row):
            if char == '@':
                robot_pos = (r, c)
            elif char == 'O':
                boxes.add((r, c))
    return robot_pos, boxes

def move_robot(warehouse, robot_pos, boxes, move):
    directions = {'^': (-1, 0), 'v': (1, 0), '<': (0, -1), '>': (0, 1)}
    dr, dc = directions[move]
    new_robot_pos = (robot_pos[0] + dr, robot_pos[1] + dc)
    
    if warehouse[new_robot_pos[0]][new_robot_pos[1]] == '#':
        return robot_pos, boxes  # Move blocked by wall
    
    if new_robot_pos in boxes:
        new_box_pos = (new_robot_pos[0] + dr, new_robot_pos[1] + dc)
        if warehouse[new_box_pos[0]][new_box_pos[1]] == '#' or new_box_pos in boxes:
            return robot_pos, boxes  # Move blocked by wall or another box
        boxes.remove(new_robot_pos)
        boxes.add(new_box_pos)
    
    return new_robot_pos, boxes

def calculate_gps_coordinates(boxes, rows, cols):
    gps_sum = 0
    for r, c in boxes:
        gps_sum += 100 * r + c
    return gps_sum

def main():
    warehouse, moves = read_input('input.txt')
    rows, cols = len(warehouse), len(warehouse[0])
    robot_pos, boxes = find_robot_and_boxes(warehouse)
    
    for move in moves:
        robot_pos, boxes = move_robot(warehouse, robot_pos, boxes, move)
    
    gps_sum = calculate_gps_coordinates(boxes, rows, cols)
    print(gps_sum)

if __name__ == "__main__":
    main()
```
