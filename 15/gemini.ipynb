```python
# Read input from 'input.txt'
with open('input.txt', 'r') as f:
    lines = f.readlines()

# Parse the map
map_lines = []
i = 0
while i < len(lines):
    line = lines[i].rstrip('\n')
    if line.startswith('#') and line.endswith('#'):
        map_lines.append(list(line))
        i += 1
    else:
        break
# Find the robot position and replace '@' with '.'
for y in range(len(map_lines)):
    for x in range(len(map_lines[y])):
        if map_lines[y][x] == '@':
            robot_y, robot_x = y, x
            map_lines[y][x] = '.'

# Read moves
moves = ''
while i < len(lines):
    line = lines[i].strip()
    moves += ''.join(c for c in line if c in '^v<>')
    i += 1

# Direction deltas
dir_map = {'^': (-1, 0), 'v': (1, 0), '<': (0, -1), '>': (0, 1)}

# Perform moves
for move in moves:
    if move not in dir_map:
        continue
    dy, dx = dir_map[move]
    ny, nx = robot_y + dy, robot_x + dx
    # Check boundaries
    if 0 <= ny < len(map_lines) and 0 <= nx < len(map_lines[ny]):
        target = map_lines[ny][nx]
        if target == '.':
            robot_y, robot_x = ny, nx
        elif target == 'O':
            by, bx = ny + dy, nx + dx
            if 0 <= by < len(map_lines) and 0 <= bx < len(map_lines[by]):
                beyond = map_lines[by][bx]
                if beyond == '.':
                    map_lines[by][bx] = 'O'
                    map_lines[ny][nx] = '.'
                    robot_y, robot_x = ny, nx

# Compute GPS coordinates
gps_sum = 0
for y in range(len(map_lines)):
    for x in range(len(map_lines[y])):
        if map_lines[y][x] == 'O':
            gps_sum += 100 * y + x

print(gps_sum)
```

```python
def solve():
    with open("input.txt", "r") as f:
        lines = f.readlines()

    grid = []
    moves = ""
    for i, line in enumerate(lines):
        line = line.strip()
        if line.startswith("#") or line.startswith(".") or line.startswith("O") or line.startswith("@"):
            grid.append(list(line))
        elif line:
            moves += line

    rows = len(grid)
    cols = len(grid[0])

    def find_robot():
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == "@":
                    return r, c
        return -1, -1

    def is_valid(r, c):
        return 0 <= r < rows and 0 <= c < cols

    def move_robot(r, c, dr, dc):
        nr, nc = r + dr, c + dc

        # If next position is invalid or a wall, robot does not move
        if not is_valid(nr, nc) or grid[nr][nc] == "#":
            return r, c

        # If next position is a box
        if grid[nr][nc] == "O":
            nnr, nnc = nr + dr, nc + dc
            # If the position after the box is invalid or a wall or another box, robot does not move
            if not is_valid(nnr, nnc) or grid[nnr][nnc] == "#" or grid[nnr][nnc] == "O":
                return r, c
            else:
                # Move the box
                grid[nnr][nnc] = "O"
                # Move the robot
                grid[nr][nc] = "@"
                grid[r][c] = "."
                return nr, nc

        # If next position is empty, move the robot
        grid[nr][nc] = "@"
        grid[r][c] = "."
        return nr, nc

    robot_r, robot_c = find_robot()

    for move in moves:
        if move == "^":
            robot_r, robot_c = move_robot(robot_r, robot_c, -1, 0)
        elif move == "v":
            robot_r, robot_c = move_robot(robot_r, robot_c, 1, 0)
        elif move == "<":
            robot_r, robot_c = move_robot(robot_r, robot_c, 0, -1)
        elif move == ">":
            robot_r, robot_c = move_robot(robot_r, robot_c, 0, 1)

    total_gps = 0
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "O":
                total_gps += (r * 100) + c

    print(total_gps)

solve()
```
