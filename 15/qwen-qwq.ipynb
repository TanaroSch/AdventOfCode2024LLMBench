```python
def main():
    with open("input.txt", "r") as f:
        lines = f.readlines()

    # Step 1: Parse the map
    map_lines = []
    i = 0
    while i < len(lines) and lines[i].strip() != "":
        map_lines.append(lines[i].strip())
        i += 1

    # Read the moves
    moves = ""
    while i < len(lines):
        moves += lines[i].strip()
        i += 1

    # Find initial positions
    boxes = set()
    for r, row in enumerate(map_lines):
        for c, char in enumerate(row):
            if char == '@':
                robot_pos = (r, c)
            elif char == 'O':
                boxes.add((r, c))

    # Define directions
    dir_moves = {
        '<': (0, -1),
        '>': (0, 1),
        '^': (-1, 0),
        'v': (1, 0)
    }

    # Simulate the moves
    for move in moves:
        dr, dc = dir_moves[move]
        nr, nc = robot_pos[0] + dr, robot_pos[1] + dc

        # Check if new position is within bounds
        if not (0 <= nr < len(map_lines) and 0 <= nc < len(map_lines[0])):
            continue

        if map_lines[nr][nc] == '#':
            # Can't move
            continue
        elif map_lines[nr][nc] == '.':
            # Move robot to empty space
            robot_pos = (nr, nc)
        elif map_lines[nr][nc] == 'O':
            # Try to push the box
            box_nr, box_nc = nr + dr, nc + dc
            # Check if box new position is within bounds
            if not (0 <= box_nr < len(map_lines) and 0 <= box_nc < len(map_lines[0])):
                continue
            if map_lines[box_nr][box_nc] == '.':
                # Move the box
                if (nr, nc) in boxes:
                    boxes.remove((nr, nc))
                    boxes.add((box_nr, box_nc))
                    # Move robot to the box's original position
                    robot_pos = (nr, nc)
                else:
                    # Box not found; cannot push
                    continue
            else:
                # Can't push the box
                continue

    # Calculate GPS sum
    gps_sum = 0
    for box in boxes:
        r, c = box
        gps = 100 * (r + 1) + (c + 1)
        gps_sum += gps

    print(gps_sum)

if __name__ == "__main__":
    main()
```
