# Day 14 - o1-preview (GHC)

```python
# Read input from input.txt
robots = []
with open('input.txt', 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        p_part, v_part = line.split()
        x_str, y_str = p_part[2:].split(',')
        dx_str, dy_str = v_part[2:].split(',')
        x = int(x_str)
        y = int(y_str)
        dx = int(dx_str)
        dy = int(dy_str)
        robots.append((x, y, dx, dy))

# Dimensions of the space
width = 101
height = 103

# Middle positions
middle_x = width // 2  # 50
middle_y = height // 2  # 51

# Counts for quadrants
quadrant_counts = [0, 0, 0, 0]  # Quadrants 1 to 4

# Simulate the robots after 100 seconds
for x, y, dx, dy in robots:
    x_final = (x + dx * 100) % width
    y_final = (y + dy * 100) % height

    if x_final == middle_x or y_final == middle_y:
        continue  # Ignore robots on the middle lines
    if x_final < middle_x and y_final < middle_y:
        quadrant_counts[0] += 1  # Quadrant 1
    elif x_final >= middle_x and y_final < middle_y:
        quadrant_counts[1] += 1  # Quadrant 2
    elif x_final < middle_x and y_final >= middle_y:
        quadrant_counts[2] += 1  # Quadrant 3
    elif x_final >= middle_x and y_final >= middle_y:
        quadrant_counts[3] += 1  # Quadrant 4

# Calculate the safety factor
safety_factor = 1
for count in quadrant_counts:
    safety_factor *= count

print(safety_factor)
```
