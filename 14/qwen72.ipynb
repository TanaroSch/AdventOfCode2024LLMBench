```python
def read_input(file_path):
    robots = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            pos = tuple(map(int, parts[0][2:-1].split(',')))
            vel = tuple(map(int, parts[1][2:-1].split(',')))
            robots.append((pos, vel))
    return robots

def simulate_movement(robots, width, height, seconds):
    new_positions = []
    for (pos, vel) in robots:
        x, y = pos
        vx, vy = vel
        x = (x + vx * seconds) % width
        y = (y + vy * seconds) % height
        new_positions.append((x, y))
    return new_positions

def count_robots_in_quadrants(robots, width, height):
    quadrant_counts = [0, 0, 0, 0]
    mid_x = width // 2
    mid_y = height // 2

    for (x, y) in robots:
        if x < mid_x and y < mid_y:
            quadrant_counts[0] += 1
        elif x >= mid_x and y < mid_y:
            quadrant_counts[1] += 1
        elif x < mid_x and y >= mid_y:
            quadrant_counts[2] += 1
        elif x >= mid_x and y >= mid_y:
            quadrant_counts[3] += 1

    return quadrant_counts

def compute_safety_factor(quadrant_counts):
    return quadrant_counts[0] * quadrant_counts[1] * quadrant_counts[2] * quadrant_counts[3]

def main():
    width = 101
    height = 103
    seconds = 100
    file_path = 'input.txt'

    robots = read_input(file_path)
    new_positions = simulate_movement(robots, width, height, seconds)
    quadrant_counts = count_robots_in_quadrants(new_positions, width, height)
    safety_factor = compute_safety_factor(quadrant_counts)

    print(f"Quadrant counts: {quadrant_counts}")
    print(f"Safety factor: {safety_factor}")

if __name__ == "__main__":
    main()
```

```python
def parse_robots(input_lines):
    robots = []
    for line in input_lines:
        p, v = line.split()
        px, py = map(int, p[2:].split(','))
        vx, vy = map(int, v[2:].split(','))
        robots.append(((px, py), (vx, vy)))
    return robots

def move_robot(position, velocity, time):
    px, py = position
    vx, vy = velocity
    new_px = (px + vx * time) % 101
    new_py = (py + vy * time) % 103
    return (new_px, new_py)

def count_robots_in_quadrants(robots):
    quadrant_counts = [0, 0, 0, 0]
    for px, py in robots:
        if px == 0 or px == 100 or py == 0 or py == 102:
            continue  # Skip robots on the edges
        if px < 51 and py < 52:
            quadrant_counts[0] += 1
        elif px >= 51 and py < 52:
            quadrant_counts[1] += 1
        elif px < 51 and py >= 52:
            quadrant_counts[2] += 1
        elif px >= 51 and py >= 52:
            quadrant_counts[3] += 1
    return quadrant_counts

def calculate_safety_factor(quadrant_counts):
    return quadrant_counts[0] * quadrant_counts[1] * quadrant_counts[2] * quadrant_counts[3]

def main():
    with open('input.txt', 'r') as file:
        input_lines = file.readlines()
    
    robots = parse_robots(input_lines)
    new_positions = [move_robot(robot[0], robot[1], 100) for robot in robots]
    quadrant_counts = count_robots_in_quadrants(new_positions)
    safety_factor = calculate_safety_factor(quadrant_counts)
    print(f"Safety factor: {safety_factor}")

if __name__ == "__main__":
    main()
```
