```python
width = 101
height = 103

# Read input data
with open("input.txt", "r") as f:
    lines = f.readlines()

robots = []
for line in lines:
    line = line.strip()
    # Example line: p=0,4 v=3,-3
    # Split by spaces
    parts = line.split()
    # parts[0] should be like p=0,4
    # parts[1] should be like v=3,-3
    p_part = parts[0][2:]  # remove 'p='
    v_part = parts[1][2:]  # remove 'v='

    # p_part: "0,4"
    # v_part: "3,-3"
    px, py = map(int, p_part.split(","))
    vx, vy = map(int, v_part.split(","))
    robots.append((px, py, vx, vy))

# Simulate after 100 seconds
time = 100
final_positions = []
for (px, py, vx, vy) in robots:
    final_x = (px + vx*time) % width
    final_y = (py + vy*time) % height
    final_positions.append((final_x, final_y))

# Count robots in each quadrant
q1 = 0
q2 = 0
q3 = 0
q4 = 0

# Quadrants definition:
# Middle lines at x=50, y=51
for (x, y) in final_positions:
    if x > 50 and y < 51:
        q1 += 1
    elif x < 50 and y < 51:
        q2 += 1
    elif x < 50 and y > 51:
        q3 += 1
    elif x > 50 and y > 51:
        q4 += 1
    # If x=50 or y=51, we ignore that robot

# Compute the safety factor
safety_factor = q1 * q2 * q3 * q4

print(safety_factor)
```

```python
def bounding_box(points):
    xs = [p[0] for p in points]
    ys = [p[1] for p in points]
    min_x, max_x = min(xs), max(xs)
    min_y, max_y = min(ys), max(ys)
    return min_x, max_x, min_y, max_y

# Read input
with open("input.txt", "r") as f:
    lines = f.readlines()

robots = []
for line in lines:
    line = line.strip()
    # Format: p=X,Y v=Dx,Dy
    parts = line.split()
    p_part = parts[0][2:]  # after 'p='
    v_part = parts[1][2:]  # after 'v='
    px, py = map(int, p_part.split(','))
    vx, vy = map(int, v_part.split(','))
    robots.append([px, py, vx, vy])

time = 0
points = [[r[0], r[1]] for r in robots]

min_area = None
min_area_time = None

# We'll simulate until we detect that we've passed the minimal area.
# Since the pattern should form at a unique minimal area moment, we can rely on that.
previous_area = None
increasing_count = 0  # count how many times in a row area increased since last minimum

while True:
    # Compute bounding box
    min_x, max_x, min_y, max_y = bounding_box(points)
    area = (max_x - min_x + 1) * (max_y - min_y + 1)

    if min_area is None or area < min_area:
        min_area = area
        min_area_time = time
        increasing_count = 0  # reset since we found a smaller area
    else:
        # If the area is larger or equal to the last recorded minimal area, it might be getting bigger
        if previous_area is not None and area > previous_area:
            increasing_count += 1
        else:
            # If it didn't increase, reset
            increasing_count = 0

        # If area is growing for several consecutive steps after the minimum,
        # we can be fairly sure we've passed the minimal configuration.
        # Usually 5 or 10 steps is enough for a puzzle like this.
        if min_area_time is not None and increasing_count >= 10:
            break

    previous_area = area

    # Move points one second forward
    for p, r in zip(points, robots):
        p[0] += r[2]  # vx
        p[1] += r[3]  # vy
    time += 1

print(min_area_time)
```

```python
import re

def bounding_box(points):
    xs = [p[0] for p in points]
    ys = [p[1] for p in points]
    return min(xs), max(xs), min(ys), max(ys)

# Read input
with open("input.txt", "r") as f:
    lines = f.readlines()

robots = []
# The AoC line format: position=<X, Y> velocity=<DX, DY>
# Use regex to parse
pattern = re.compile(r"position=<\s*(-?\d+),\s*(-?\d+)>\s*velocity=<\s*(-?\d+),\s*(-?\d+)>")

for line in lines:
    line = line.strip()
    m = pattern.match(line)
    if m:
        px, py, vx, vy = map(int, m.groups())
        # Store position and velocity
        robots.append([px, py, vx, vy])
    else:
        # If no match, input line might not be in AoC format
        # You can raise an error or print a warning
        print("Line not in expected format:", line)

# Positions at time=0
points = [[r[0], r[1]] for r in robots]

min_area = None
min_area_time = None

MAX_TIME = 20000  # Should be enough for the AoC 2018 Day 10 puzzle

for t in range(MAX_TIME + 1):
    # Compute bounding box
    min_x, max_x, min_y, max_y = bounding_box(points)
    width = (max_x - min_x + 1)
    height = (max_y - min_y + 1)
    area = width * height
    
    if min_area is None or area < min_area:
        min_area = area
        min_area_time = t
    
    # Move points forward for next second
    for i in range(len(points)):
        px, py = points[i]
        vx, vy = robots[i][2], robots[i][3]
        points[i][0] = px + vx
        points[i][1] = py + vy

print(min_area_time)
```
