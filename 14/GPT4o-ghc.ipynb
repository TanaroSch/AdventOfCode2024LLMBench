```python
# Define the dimensions of the space
width = 101
height = 103

# Read the input file
with open('input.txt', 'r') as file:
    lines = file.readlines()

# Parse the input to get positions and velocities
robots = []
for line in lines:
    parts = line.strip().split()
    p = parts[0][2:].split(',')
    v = parts[1][2:].split(',')
    position = (int(p[0]), int(p[1]))
    velocity = (int(v[0]), int(v[1]))
    robots.append((position, velocity))

# Simulate the movement for 100 seconds
for _ in range(100):
    new_robots = []
    for position, velocity in robots:
        new_x = (position[0] + velocity[0]) % width
        new_y = (position[1] + velocity[1]) % height
        new_robots.append(((new_x, new_y), velocity))
    robots = new_robots

# Count the number of robots in each quadrant
quadrant_counts = [0, 0, 0, 0]
for position, _ in robots:
    x, y = position
    if x == width // 2 or y == height // 2:
        continue
    if x < width // 2 and y < height // 2:
        quadrant_counts[0] += 1
    elif x >= width // 2 and y < height // 2:
        quadrant_counts[1] += 1
    elif x < width // 2 and y >= height // 2:
        quadrant_counts[2] += 1
    elif x >= width // 2 and y >= height // 2:
        quadrant_counts[3] += 1

# Calculate the safety factor
safety_factor = quadrant_counts[0] * quadrant_counts[1] * quadrant_counts[2] * quadrant_counts[3]

print(f'Safety Factor: {safety_factor}')
```
