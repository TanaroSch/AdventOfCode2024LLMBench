```python
def read_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    # Parse the initial values of the registers
    registers = {}
    for line in lines[:3]:
        key, value = line.strip().split(': ')
        registers[key.split()[1]] = int(value)
    
    # Parse the program
    program_line = lines[4].split(': ')[1]
    program = list(map(int, program_line.split(',')))
    
    return registers, program

def execute_program(registers, program):
    # Opcodes and their corresponding actions
    def adv(operand):
        denominator = 2 ** (operand if operand < 4 else registers['B'])
        registers['A'] = registers['A'] // denominator
    
    def bxl(operand):
        registers['B'] = registers['B'] ^ operand
    
    def bst(operand):
        registers['B'] = operand % 8
    
    def jnz(operand):
        if registers['A'] != 0:
            return operand
        return None
    
    def bxc(operand):
        registers['B'] = registers['B'] ^ registers['C']
    
    def out(operand):
        value = operand % 8
        output.append(value)
    
    def bdv(operand):
        denominator = 2 ** (operand if operand < 4 else registers['B'])
        registers['B'] = registers['A'] // denominator
    
    def cdv(operand):
        denominator = 2 ** (operand if operand < 4 else registers['B'])
        registers['C'] = registers['A'] // denominator
    
    # Mapping opcodes to functions
    opcode_functions = {
        0: adv,
        1: bxl,
        2: bst,
        3: jnz,
        4: bxc,
        5: out,
        6: bdv,
        7: cdv
    }
    
    # Execute the program
    output = []
    ip = 0
    while ip < len(program):
        opcode = program[ip]
        operand = program[ip + 1]
        ip += 2
        
        if opcode == 3:
            jump_ip = jnz(operand)
            if jump_ip is not None:
                ip = jump_ip
        else:
            opcode_functions[opcode](operand)
    
    return output

def main():
    registers, program = read_input('input.txt')
    output = execute_program(registers, program)
    print(','.join(map(str, output)))

if __name__ == '__main__':
    main()
```
