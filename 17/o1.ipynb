```python
def combo_value(operand, A, B, C):
    # combo operand: 0-3 = literal, 4=A, 5=B, 6=C
    if operand <= 3:
        return operand
    elif operand == 4:
        return A
    elif operand == 5:
        return B
    elif operand == 6:
        return C
    else:
        # operand=7 is reserved, should not appear
        raise ValueError("Invalid combo operand encountered")

def run_program(A_init, B_init, C_init, program):
    A, B, C = A_init, B_init, C_init
    IP = 0
    output_values = []

    while IP < len(program):
        opcode = program[IP]
        if IP+1 < len(program):
            operand = program[IP+1]
        else:
            # No operand available means end of program
            break

        if opcode == 0:  # adv
            # A = floor(A / (2^(combo_value(operand))))
            val = combo_value(operand, A, B, C)
            denom = 2 ** val
            A = A // denom
            IP += 2

        elif opcode == 1:  # bxl
            # B = B XOR operand (literal)
            B = B ^ operand
            IP += 2

        elif opcode == 2:  # bst
            # B = combo_value(operand) % 8
            B = combo_value(operand, A, B, C) % 8
            IP += 2

        elif opcode == 3:  # jnz
            # if A != 0, IP = operand (literal); else IP +=2
            if A != 0:
                IP = operand
            else:
                IP += 2

        elif opcode == 4:  # bxc
            # B = B XOR C, operand ignored
            B = B ^ C
            IP += 2

        elif opcode == 5:  # out
            # output combo_value(operand) % 8
            val = combo_value(operand, A, B, C) % 8
            output_values.append(val)
            IP += 2

        elif opcode == 6:  # bdv
            # B = floor(A / (2^(combo_value(operand))))
            val = combo_value(operand, A, B, C)
            B = A // (2 ** val)
            IP += 2

        elif opcode == 7:  # cdv
            # C = floor(A / (2^(combo_value(operand))))
            val = combo_value(operand, A, B, C)
            C = A // (2 ** val)
            IP += 2

        else:
            # Invalid opcode
            break

    return output_values

if __name__ == "__main__":
    # Read input
    with open('input.txt') as f:
        lines = f.read().strip().split('\n')

    # For the demonstration given in the problem:
    # Registers:
    A = 729
    B = 0
    C = 0
    # Program:
    program = [0, 1, 5, 4, 3, 0]

    outputs = run_program(A, B, C, program)
    print(",".join(map(str, outputs)))
```

```python
# Given:
program = [0,3,5,4,3,0]

# The length of the program is n
n = len(program)

# We know the output sequence O should match the program.
# O[i-1] = (A // (8^i)) % 8 for i in 1..n

# Let's denote:
# After i iterations: O[i-1] = floor(A / (8^i)) % 8

# We want to find A such that:
# floor(A/8) %8 = program[0]
# floor(A/64)%8 = program[1]
# ...
# floor(A/(8^n))=0

O = program
A = 0
# We'll reconstruct A starting from the highest iteration:
# O[n-1] = floor(A/(8^n))%8 = must be consistent with A<8^n, so floor(A/(8^n))=0 is okay.
# Work downward:
# O[n-1] gives no restriction other than A<8^n.

# For i in reverse from n-1 down to 0:
# floor(A/(8^(i+1)))%8 = O[i]
# Let X_i = floor(A/(8^(i+1))).
# X_i %8 = O[i].

# Also X_{i} = floor(A/(8^(i+1)))
# X_{i-1} = floor(A/(8^i)) = floor(X_i/8).

# We can solve iteratively:
x = 0
# We know x = floor(A/(8^n))=0 initially.
for i in reversed(range(n)):
    # O[i] = (floor(A/(8^(i+1)))%8)
    # But floor(A/(8^(i+1))) = x at iteration i.
    # We know x%8 = O[i].
    # Let x = 8*k + O[i].
    x = 8*x + O[i]

# Now, after finishing, x represents floor(A/8) since we built it backward.
# But we need to be careful: we constructed x starting from the top. Actually, we ended with x as floor(A/8).

# Wait, we derived a direct formula: working from top to bottom:
# Start with x=0 for iteration n (top), each iteration: x = 8*x + O[i].
# After processing all O[i], we get x = floor(A/8).

# Therefore A = 8*x since from the first iteration we had floor(A/8)=x and floor(A/8)%8=O[0].
A = 8*x

print(A)  # This should print 117440 for the given program.
```

```python
def read_input(filename):
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    # Parse registers
    # Expected format:
    # Register A: <value>
    # Register B: <value>
    # Register C: <value>
    # Program: <comma-separated values>
    A_line = lines[0]
    B_line = lines[1]
    C_line = lines[2]
    prog_line = lines[3]

    # Extract register values (we'll ignore A's input value for this part)
    B_val = int(B_line.split(":")[1].strip())
    C_val = int(C_line.split(":")[1].strip())

    prog_str = prog_line.split(":", 1)[1].strip()
    program = list(map(int, prog_str.split(",")))

    return B_val, C_val, program

def find_initial_A(program):
    # program = [p0, p1, p2, ..., p_{n-1}]
    # We are reconstructing A from the output pattern described above.
    # Check that pattern matches what we expect (adv 3, out 4, jnz 0 repeated):
    # This is optional but let's trust the puzzle that it is correct.

    # Reconstruct A:
    # A = 8 * X where X is formed by reversing the program and interpreting as base-8 digits:
    X = 0
    for p in reversed(program):
        X = X * 8 + p
    A = 8 * X
    return A

if __name__ == "__main__":
    B_val, C_val, program = read_input("input.txt")
    A_val = find_initial_A(program)

    # Print the lowest positive A. If A_val=0, we might need to check next positive integer,
    # but given the puzzle, the reconstructed A_val should be positive.
    # If A_val is 0, we can try incrementing until positive, but usually it won't be.
    if A_val <= 0:
        # If by any chance A is not positive, try multiples:
        # The puzzle states "lowest positive initial value", so increment if needed.
        # But in practice, the derived A from the pattern should be positive.
        candidate = 1
        while True:
            # If the puzzle scenario ensures a positive solution from the formula,
            # we might never reach here. This is a fallback.
            if candidate == A_val:
                break
            candidate += 1
        A_val = candidate
    
    print(A_val)
```

```python
def parse_register_value(line):
    # line like "Register A: 2024"
    return int(line.split(":")[1].strip())

def parse_program(line):
    # line like "Program: 0,3,5,4,3,0"
    prog_str = line.split(":", 1)[1].strip()
    return list(map(int, prog_str.split(",")))

if __name__ == "__main__":
    with open("input.txt") as f:
        lines = [l.strip() for l in f if l.strip()]

    # Parse input
    # Note: A from input is ignored for part two
    B_val = parse_register_value(lines[1])
    C_val = parse_register_value(lines[2])
    program = parse_program(lines[3])
    
    # The program is self-printing. Let p_i = program[i].
    # Reconstruct A:
    # A//8 = sum of p_i * 8^i
    X = 0
    for i, p in enumerate(program):
        X += p * (8 ** i)
    A_val = 8 * X

    # Print the found A value
    print(A_val)
```

```python
def combo_value(operand, A, B, C):
    if operand <= 3:
        return operand
    elif operand == 4:
        return A
    elif operand == 5:
        return B
    elif operand == 6:
        return C
    else:
        raise ValueError("Invalid combo operand")

def run_program(A_init, B_init, C_init, program):
    A, B, C = A_init, B_init, C_init
    IP = 0
    output_values = []

    while IP < len(program):
        opcode = program[IP]
        if IP+1 < len(program):
            operand = program[IP+1]
        else:
            break

        if opcode == 0:  # adv
            val = combo_value(operand, A, B, C)
            A = A // (2 ** val)
            IP += 2

        elif opcode == 1:  # bxl
            # literal operand
            B = B ^ operand
            IP += 2

        elif opcode == 2:  # bst
            B = combo_value(operand, A, B, C) % 8
            IP += 2

        elif opcode == 3:  # jnz
            # literal operand
            if A != 0:
                IP = operand
            else:
                IP += 2

        elif opcode == 4:  # bxc
            B = B ^ C
            IP += 2

        elif opcode == 5:  # out
            val = combo_value(operand, A, B, C) % 8
            output_values.append(val)
            IP += 2

        elif opcode == 6:  # bdv
            val = combo_value(operand, A, B, C)
            B = A // (2 ** val)
            IP += 2

        elif opcode == 7:  # cdv
            val = combo_value(operand, A, B, C)
            C = A // (2 ** val)
            IP += 2
        else:
            break

    return output_values

def compare_outputs(out_prog, target_prog):
    # Compare lexicographically:
    # return -1 if out_prog < target_prog
    # return 0 if equal
    # return 1 if out_prog > target_prog
    for o, t in zip(out_prog, target_prog):
        if o < t:
            return -1
        elif o > t:
            return 1
    # If all matched so far, compare lengths:
    if len(out_prog) < len(target_prog):
        return -1
    elif len(out_prog) > len(target_prog):
        return 1
    else:
        return 0

if __name__ == "__main__":
    # From input.txt:
    # Register A: 63281501 (ignored)
    # Register B: 0
    # Register C: 0
    # Program: 2,4,1,5,7,5,4,5,0,3,1,6,5,5,3,0
    program = [2,4,1,5,7,5,4,5,0,3,1,6,5,5,3,0]
    B_val = 0
    C_val = 0

    # Binary search for A
    low = 1
    high = 8**16  # upper bound
    answer = None

    while low <= high:
        mid = (low+high)//2
        out_vals = run_program(mid, B_val, C_val, program)
        cmp_res = compare_outputs(out_vals, program)
        if cmp_res == 0:
            # Perfect match
            answer = mid
            # Try to find smaller A
            high = mid - 1
        elif cmp_res < 0:
            # out_prog < program => need bigger A to produce larger or more correct output
            low = mid + 1
        else:
            # out_prog > program => need smaller A
            high = mid - 1

    print(answer)
```
