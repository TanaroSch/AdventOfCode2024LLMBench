```python
def run_program(program, registers):
    instruction_pointer = 0
    output = []
    while instruction_pointer < len(program):
        opcode = program[instruction_pointer]
        operand = program[instruction_pointer + 1]
        if opcode == 0:  # adv
            if operand <= 3:
                divisor = 2 ** operand
            else:
                divisor = 2 ** registers[operand - 4]
            registers[0] = registers[0] // divisor
        elif opcode == 1:  # bxl
            registers[1] = registers[1] ^ operand
        elif opcode == 2:  # bst
            if operand <= 3:
                registers[1] = operand
            else:
                registers[1] = registers[operand - 4] % 8
        elif opcode == 3:  # jnz
            if registers[0]!= 0:
                instruction_pointer = operand
                continue
        elif opcode == 4:  # bxc
            registers[1] = registers[1] ^ registers[2]
        elif opcode == 5:  # out
            if operand <= 3:
                output.append(str(operand))
            else:
                output.append(str(registers[operand - 4] % 8))
        elif opcode == 6:  # bdv
            if operand <= 3:
                divisor = 2 ** operand
            else:
                divisor = 2 ** registers[operand - 4]
            registers[1] = registers[0] // divisor
        elif opcode == 7:  # cdv
            if operand <= 3:
                divisor = 2 ** operand
            else:
                divisor = 2 ** registers[operand - 4]
            registers[2] = registers[0] // divisor
        instruction_pointer += 2
    return output

def main():
    with open('input.txt', 'r') as f:
        lines = f.readlines()
    registers = [int(lines[0].split(': ')[1]), 0, 0]
    program = list(map(int, lines[2].split(', ')))
    output = run_program(program, registers)
    print(','.join(output))

if __name__ == '__main__':
    main()
```
