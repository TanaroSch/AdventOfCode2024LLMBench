```python
def read_input(file_path):
    with open(file_path, 'r') as file:
        lines = file.read().strip().split('\n')
    
    # Parse register values
    reg_a = int(lines[0].split(': ')[1])
    reg_b = int(lines[1].split(': ')[1])
    reg_c = int(lines[2].split(': ')[1])
    
    # Parse program
    program_line = lines[3].strip().startswith('Program: ')
    program = list(map(int, lines[3].split(':')[1].strip().split(',')))
    
    return reg_a, reg_b, reg_c, program

def execute_program(reg_a, reg_b, reg_c, program):
    output_values = []
    ip = 0
    
    while ip < len(program):
        opcode = program[ip]
        operand = program[ip + 1] if ip + 1 < len(program) else None
        
        if opcode == 0:  # adv
            divisor = 2 ** get_combo_value(operand)
            reg_a //= divisor
        elif opcode == 1:  # bxl
            reg_b ^= operand
        elif opcode == 2:  # bst
            reg_b = get_combo_value(operand)
        elif opcode == 3:  # jnz
            if reg_a != 0:
                ip = operand
                continue
        elif opcode == 4:  # bxc
            reg_b ^= reg_c
        elif opcode == 5:  # out
            output_values.append(get_combo_value(operand))
        elif opcode == 6:  # bdv
            divisor = 2 ** get_combo_value(operand)
            reg_b = reg_a // divisor
        elif opcode == 7:  # cdv
            divisor = 2 ** get_combo_value(operand)
            reg_c = reg_a // divisor
        
        ip += 2
    
    return output_values

def get_combo_value(combo):
    if 0 <= combo <= 3:
        return combo
    elif combo == 4:
        return reg_a
    elif combo == 5:
        return reg_b
    elif combo == 6:
        return reg_c
    else:
        raise ValueError("Invalid combo operand")

def main():
    reg_a, reg_b, reg_c, program = read_input('input.txt')
    output_values = execute_program(reg_a, reg_b, reg_c, program)
    result = ','.join(map(str, output_values))
    print(result)

if __name__ == "__main__":
    main()
```
