```python
# Example content:
# Register A: 729
# Register B: 0
# Register C: 0
# Program: 0,1,5,4,3,0

def read_input(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    registers = {}
    for line in lines[:3]:
        parts = line.strip().split(': ')
        registers[parts[0].split()[1]] = int(parts[1])
    
    program = list(map(int, lines[3].strip().split(': ')[1].split(',')))
    
    return registers, program

def get_combo_value(operand, registers):
    if operand <= 3:
        return operand
    elif operand == 4:
        return registers['A']
    elif operand == 5:
        return registers['B']
    elif operand == 6:
        return registers['C']
    else:
        raise ValueError("Invalid combo operand")

def execute_program(registers, program):
    ip = 0
    output = []
    
    while ip < len(program):
        opcode = program[ip]
        operand = program[ip + 1]
        
        if opcode == 0:  # adv
            registers['A'] //= 2 ** get_combo_value(operand, registers)
        elif opcode == 1:  # bxl
            registers['B'] ^= operand
        elif opcode == 2:  # bst
            registers['B'] = get_combo_value(operand, registers) % 8
        elif opcode == 3:  # jnz
            if registers['A'] != 0:
                ip = operand
                continue
        elif opcode == 4:  # bxc
            registers['B'] ^= registers['C']
        elif opcode == 5:  # out
            output.append(get_combo_value(operand, registers) % 8)
        elif opcode == 6:  # bdv
            registers['B'] = registers['A'] // (2 ** get_combo_value(operand, registers))
        elif opcode == 7:  # cdv
            registers['C'] = registers['A'] // (2 ** get_combo_value(operand, registers))
        else:
            raise ValueError("Invalid opcode")
        
        ip += 2
    
    return output

def main():
    registers, program = read_input('input.txt')
    output = execute_program(registers, program)
    print(','.join(map(str, output)))

if __name__ == "__main__":
    main()
```
