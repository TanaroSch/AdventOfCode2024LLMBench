```python
with open('input.txt', 'r') as f:
    lines = f.readlines()

regs = {
    'A': int(lines[0].split(': ')[1]),
    'B': int(lines[1].split(': ')[1]),
    'C': int(lines[2].split(': ')[1]),
}

program = list(map(int, lines[3].split(': ')[1].split(',')))

class Computer:
    def __init__(self, A, B, C, program):
        self.A = A
        self.B = B
        self.C = C
        self.IP = 0
        self.program = program
        self.output = []

    def get_combo_operand(self, operand):
        if operand <= 3:
            return operand
        elif operand == 4:
            return self.A
        elif operand == 5:
            return self.B
        elif operand == 6:
            return self.C
        elif operand == 7:
            raise ValueError("Reserved operand 7 used")

    def run(self):
        while self.IP < len(self.program):
            opcode = self.program[self.IP]
            operand = self.program[self.IP + 1]
            if opcode == 0:  # adv
                value = self.get_combo_operand(operand)
                self.A = self.A // (2 ** value)
            elif opcode == 1:  # bxl
                value = operand  # literal operand
                self.B = self.B ^ value
            elif opcode == 2:  # bst
                value = self.get_combo_operand(operand)
                self.B = value % 8
            elif opcode == 3:  # jnz
                value = operand  # literal operand
                if self.A!= 0:
                    self.IP = value - 2
            elif opcode == 4:  # bxc
                # ignore operand
                self.B = self.B ^ self.C
            elif opcode == 5:  # out
                value = self.get_combo_operand(operand)
                self.output.append(str(value % 8))
            elif opcode == 6:  # bdv
                value = self.get_combo_operand(operand)
                self.B = self.A // (2 ** value)
            elif opcode == 7:  # cdv
                value = self.get_combo_operand(operand)
                self.C = self.A // (2 ** value)
            else:
                raise ValueError(f"Invalid opcode: {opcode}")
            # Increment IP by 2, but not if we just jumped
            if opcode!= 3 or self.A == 0:
                self.IP += 2

    def get_output(self):
        return ','.join(self.output)

computer = Computer(regs['A'], regs['B'], regs['C'], program)
computer.run()
print(computer.get_output())
```
