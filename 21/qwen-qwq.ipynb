```python
from collections import deque

# Define positions for numeric keypad
numeric_positions = {
    '7': (0,0), '8': (0,1), '9': (0,2),
    '4': (1,0), '5': (1,1), '6': (1,2),
    '1': (2,0), '2': (2,1), '3': (2,2),
    '0': (3,1), 'A': (3,2)
}

# Define positions for directional keypad
directional_positions = {
    '^': (0,0), 'A': (0,1),
    '<': (1,0), 'v': (1,1), '>': (1,2)
}

# Define numeric keypad layout
numeric_keypad = [
    ['7', '8', '9'],
    ['4', '5', '6'],
    ['1', '2', '3'],
    ['0', 'A']
]

# Define directional keypad layout
directional_keypad = [
    ['^', 'A'],
    ['<', 'v', '>']
]

def get_distance(keypad, positions, start, end):
    """Get the distance between two buttons on the keypad."""
    start_pos = positions[start]
    end_pos = positions[end]
    queue = deque([(start_pos, 0)])
    visited = set()
    while queue:
        (x, y), dist = queue.popleft()
        if (x, y) == end_pos:
            return dist
        if (x, y) in visited:
            continue
        visited.add((x, y))
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < len(keypad):
                row_length = len(keypad[nx])
                if 0 <= ny < row_length and keypad[nx][ny] != '':
                    queue.append(((nx, ny), dist + 1))
    return -1

def calculate_sequence_length(code):
    """Calculate the sequence length for a given code."""
    current = 'A'  # Starting position
    total_distance = 0
    for char in code:
        if char == 'A':
            # Special handling for 'A' as it might require different movement
            distance = get_distance(numeric_keypad, numeric_positions, current, 'A')
            total_distance += distance + 1  # +1 for activation
        else:
            distance = get_distance(numeric_keypad, numeric_positions, current, char)
            total_distance += distance + 1  # +1 for activation
        current = char
    return total_distance

def main():
    sequences = ['029A', '980A', '179A', '456A', '379A']
    total_complexity = 0
    for sequence in sequences:
        sequence_length = calculate_sequence_length(sequence)
        numeric_part = int(''.join(filter(str.isdigit, sequence)))
        complexity = sequence_length * numeric_part
        total_complexity += complexity
    print(total_complexity)

if __name__ == "__main__":
    main()
```
